Data Analysis 4: t- test of realtionship between ablility to read and write and sentence length

```python
stats.levene(read_sentence['Cannot read or write'], read_sentence['Can Read and Write'])


stats.shapiro(read_sentence['Cannot read or write'])

stats.shapiro(read_sentence['Can Read and Write'])


stats.ttest_ind(read_sentence['Cannot read or write'], read_sentence['Can Read and Write'])
```
